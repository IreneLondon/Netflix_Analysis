#### Importing Libraries 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker
import matplotlib.cm as cm
import matplotlib as mpl
from matplotlib.gridspec import GridSpec
```

### EDA

#### Cleaning Data

```python
netflix_df.info()
```

```python
netflix_df.columns
```

```python
# removing columns not needed for this analysis
net_df = netflix_df.drop(columns = ['Summary','Image','Writer','Director','TMDb Trailer','Poster','Trailer Site','Actors','Production House','Netflix Link','IMDb Link'])
net_df.head(5)
```

```python
net_df.describe()
```

```python
# finding nan values
net_df[net_df.isna().any(axis = 1)]
```

```python
# replacing nan values with 0
net_df = net_df.replace(np.nan,0)
net_df
```

### Adding new columns

```python
# Converting Release date into month and year columns
net_df['Release_Month'] = pd.DatetimeIndex(net_df['Release Date']).month
net_df['Release_Year'] = pd.DatetimeIndex(net_df['Release Date']).year
```

```python
# Converting Release date into month and year columns
net_df['Netflix_Release_Month'] = pd.DatetimeIndex(net_df['Netflix Release Date']).month
net_df['Netflix_Release_Year'] = pd.DatetimeIndex(net_df['Netflix Release Date']).year
```

```python
# Confirming new columns added
net_df.info()
```
